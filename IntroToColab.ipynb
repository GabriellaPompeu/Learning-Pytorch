# **TENSORS**
*   A torch.Tensor is a multi-dimensional matrix containing elements of a single data type.
*   default data type of float32.
* is more suitable for deep learning than a numpy array.



```python
import torch
import numpy as np
```

```python
# Create a tensor 2-D with random values
tensor_2d = torch.randn(3, 4)
tensor_2d
```

```python
# Create a tensor 3-D with zeros
tensor_3d = torch.zeros(2, 3, 4)
tensor_3d
```

```python
my_torch = torch.arange(10)
my_torch
```

```python
# Reshape and View
my_torch = my_torch.reshape(2, 5)
my_torch
```

```python
# Reshape if we don't know the number of items using -1
my_torch2 = torch.arange(14)
my_torch2 = my_torch2.reshape(2, -1)
my_torch2
```

```python
my_torch3 = torch.randn(10)
my_torch3 = my_torch3.reshape(2, 5)
my_torch3
```

```python
tensor_res = torch.add(my_torch, my_torch3)
tensor_res
```

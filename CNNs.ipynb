#Convolutional Neural Network
##Definition:
* Is a specialized type of deep learning model to process and analyze visual data.
* Automatically learn features through convolutional operations.
##Key Components:
* Input Layer: receives the raw image data and passes it to the network for processing.
* Convolutional Layer: Applies a set of kernels and extracts important features from the input.
* Activation Layer: Introduces non-linearity into the cnn.
